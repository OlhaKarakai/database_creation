```python
import sqlite3
import csv

# З'єднання з базою даних (створення або підключення)
conn = sqlite3.connect('better_me.db')
cursor = conn.cursor()

# Створення таблиць, якщо їх ще немає
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Event (
        EventID INTEGER PRIMARY KEY,
        EventDate TEXT,
        AppName TEXT,
        AppAppleID INTEGER,
        SubscriptionAppleID INTEGER,
        FOREIGN KEY (SubscriptionAppleID) references Subscription (SubscriptionAppleID)
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Subscription (
        SubscriptionID INTEGER PRIMARY KEY,
        SubscriptionName TEXT,
        SubscriptionAppleID INTEGER,
        SubscriptionGroupID INTEGER,
        SubscriptionDuration TEXT,
        FOREIGN KEY (SubscriptionAppleID) references Purchase (SubscriptionAppleID)
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Purchase (
        PurchaseID INTEGER PRIMARY KEY,
        SubscriberID REAL,
        SubscriptionAppleID INTEGER,
        Units INTEGER,
        CustomerPrice REAL,
        CustomerCurrency TEXT,
        DeveloperProceeds REAL,
        ProceedsCurrency TEXT,
        IntroductoryPriceType TEXT,
        IntroductoryPriceDuration TEXT,
        MarketingOptInDuration TEXT,
        PreservedPricing TEXT,
        ProceedsReason TEXT,
        Refund TEXT,
        PurchaseDate TEXT,
        SubscriberIDReset TEXT,
        Client TEXT,
        FOREIGN KEY (SubscriberID) references User (SubscriberID)
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS User (
        UserID INTEGER PRIMARY KEY,
        SubscriberID REAL,
        Device TEXT,
        Country TEXT,
        FOREIGN KEY (SubscriberID) references Purchase (SubscriberID)
    )
''')

# Зчитування даних з текстового файлу та їх вставка в базу даних
with open('textfile.txt', 'r') as file:
    data = csv.reader(file, delimiter='\t')
    next(data)  # Пропустити заголовок

    for row in data:
        cursor.execute('INSERT INTO Event VALUES (NULL, ?, ?, ?, ?)', (row[0], row[1], row[2], row[4]))
        cursor.execute('INSERT INTO Subscription VALUES (NULL, ?, ?, ?, ?)', (row[3], row[4], row[5], row[6]))
        cursor.execute('INSERT INTO Purchase VALUES (NULL, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)', 
                       (row[17], row[4], row[23], row[10], row[11], row[12], row[13], row[7], row[8], row[9], row[14], row[15], row[21], row[22], row[20], row[16]))
        cursor.execute('INSERT INTO User VALUES (NULL, ?, ?, ?)', (row[17], row[18], row[19]))

# Збереження змін та закриття з'єднання
conn.commit()
conn.close()
```
